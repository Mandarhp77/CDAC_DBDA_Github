```python
import h2o
from h2o.estimators.random_forest import H2ORandomForestEstimator
from h2o.estimators.gbm import H2OGradientBoostingEstimator
from h2o.estimators.glm import H2OGeneralizedLinearEstimator
```

```python
h2o.init()
```

```python
df = h2o.import_file("C:\Aalesh and Mandar\Datasets\Bankruptcy.csv")
```

```python
df.col_names
```

```python
all_cols = df.col_names
```

```python
y='D'
x = all_cols[3:]
```

```python
df['D'] = df['D'].asfactor()
print(df['D'].levels())
```

```python
train,test = df.split_frame(ratios=[.7])
print(df.shape)
print(train.shape)
print(test.shape)
```

```python
h2o_rf = H2ORandomForestEstimator(seed=2022)
h2o_rf.train(x=x,y=y,training_frame=train,validation_frame=test,model_id="h2o_rf_cancer")
```

```python
print(h2o_rf.auc())
print(h2o_rf.confusion_matrix())

y_pred = h2o_rf.predict(test_data=test)
y_pred_df = y_pred.as_data_frame()

h2o_rf.logloss(train=False,valid=True,xval=False)
```

```python
h2o_gb = H2OGradientBoostingEstimator(seed=2022)
h2o_gb.train(x=x,y=y,training_frame=train,validation_frame=test,model_id="h2o_rf_cancer")
```

```python
print(h2o_gb.auc())
print(h2o_gb.confusion_matrix())

y_pred = h2o_gb.predict(test_data=test)
y_pred_df = y_pred.as_data_frame()

h2o_gb.logloss(train=False,valid=True,xval=False)
```

```python
h2o_le =  H2OGeneralizedLinearEstimator(seed=2022)
h2o_le.train(x=x,y=y,training_frame=train,validation_frame=test,model_id="h2o_rf_cancer")
```

```python
print(h2o_le.auc())
print(h2o_le.confusion_matrix())

y_pred = h2o_le.predict(test_data=test)
y_pred_df = y_pred.as_data_frame()

h2o_gb.logloss(train=False,valid=True,xval=False)
```
