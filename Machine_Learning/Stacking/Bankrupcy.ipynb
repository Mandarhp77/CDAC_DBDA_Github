```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.ensemble import StackingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, StratifiedKFold
```

```python
bank = pd.read_csv("C:\Aalesh and Mandar\Datasets\Cases\Bankruptcy\Bankruptcy.csv")
```

```python
bank.head()
```

```python
x = bank.drop(["D","YR","NO"], axis=1)
y = bank["D"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, stratify = y, random_state = 2022, test_size = 0.3)
```

```python
lr = LogisticRegression()
rfc = RandomForestClassifier()
nb = GaussianNB()
lda = LinearDiscriminantAnalysis()
stack = StackingClassifier([("LR", lr),("NB",nb),("LDA",lda)],final_estimator=rfc,passthrough=True)
```

```python
stack.fit(x_train, y_train)
y_pred = stack.predict(x_test)
y_pred_prob = stack.predict_proba(x_test)[:,1]
print(accuracy_score(y_test, y_pred))
print(roc_auc_score(y_test, y_pred_prob))
```

```python
stack.get_params()
```

```python
kfold = StratifiedKFold(n_splits = 5, shuffle = True, random_state = 2022)
params = {"LR__C": np.linspace(0,5,5),
          "final_estimator__max_features":[2,4,6,8]}
gcv = GridSearchCV(stack, param_grid = params, verbose = 3, scoring = "roc_auc", cv = kfold)
gcv.fit(x, y)
```

```python
print(gcv.best_params_)
print(gcv.best_score_)
```
